```python
def costfunction(theta0,theta1,data):
    totalerror=0
    for i in range(0,len(data)):
        y=data[i,1]
        x=data[i,0]
        totalerror+=(y-(theta0+theta1*x)-y)**2
    return (totalerror/float(len(data)))
    
    
```

```python
def gradientdescent(theta0,theta1,data,learningrate):
    derivativetheta0=0
    derivativetheta1=0
    for i in range(len(data)):
        y=data[i,1]
        x=data[i,0]
        derivativetheta0+=-(2*(y-(theta0+theta1*x)))/len(data)
        derivativetheta1+=-2*x*(y-(theta0+theta1*x))/len(data)
    theta0=theta0-(learningrate*derivativetheta0)
    theta1=theta1-(learningrate*derivativetheta1)
    return[theta0,theta1]
        
        
```

```python
def gradientdescentconverge(theta0,theta1,learningrate,iterations):
    
    for i in range(0,iterations):
        
        [theta0,theta1]=gradientdescent(theta0,theta1,data,learningrate)
        print("theta0",theta0,"theta1",theta1,"error___",costfunction(theta0,theta1,data))
        
    return [theta0,theta1]
        
```
